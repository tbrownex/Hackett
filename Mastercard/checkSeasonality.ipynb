#### Periodogram

```python
from config import getClient
import matplotlib.pyplot as plt
import math
import pandas as pd
from scipy import signal
import numpy as np
```

```python
dataloc = getClient("MC")
```

```python
df = pd.read_csv(dataloc+"UAEseries.csv")
```

```python
A1 = .866          # Amplitudes
A2 = .5
F1 = 12            # Monthly
F2 = 52            # Weekly
N  = 10000         # Number of observations
INTERVAL = 10000   # observations/second
```

```python
plt.plot(df["Amount"])
plt.show()
```

```python
f, Pxx_den = signal.periodogram(df["Amount"])
plt.semilogy(f[:60], Pxx_den[:60])
plt.ylim([1e-7, 1e2])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
