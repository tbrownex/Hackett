##### Show the Banstead sum of sales before and after imputation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PATH = '/home/tbrownex/data/Hackett/JLP/'
F1   = '4 Banstead.csv'
F2   = '5 Banstead_imputed.csv'
```

```python
def printSales(data):
    from matplotlib.ticker import MultipleLocator, FormatStrFormatter
    majorLocator = MultipleLocator(5)
    majorFormatter = FormatStrFormatter('%d')
    minorLocator = MultipleLocator(1)
    
    fig, ax = plt.subplots()
    
    plt.plot(data)
    plt.axvline(x=24,color='r', linewidth=.75)
    plt.axvline(x=28,color='r', linewidth=.75)
    ax.set_ylim([200000,data.max()])

    ax.xaxis.set_major_locator(majorLocator)
    ax.xaxis.set_major_formatter(majorFormatter)
# for the minor ticks, use no labels; default NullFormatter
    ax.xaxis.set_minor_locator(minorLocator)
    plt.show()
```

```python
before = pd.read_csv(PATH+F1)
after  = pd.read_csv(PATH+F2)
```

```python
before = before.set_index("Trading Week")
before = before.groupby("Trading Week")["Full Price Sales"].sum()
before = before.loc[before.index > "2017"]

printSales(before)
```

```python
after = after.set_index("Trading Week")
after = after.groupby("Trading Week")["Full Price Sales"].sum()
after = after.loc[after.index > "2017"]

printSales(after)
```

```python
dfList = [before, after]
for df in dfList:
    df = df.set_index("Trading Week")
    df = df.groupby("Trading Week")["Full Price Sales"].sum()
    df = df.loc[df.index > "2017"]
    printSales(df)
```
