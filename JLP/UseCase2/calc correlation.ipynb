```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

PATH = '/home/tbrownex/data/Hackett/JLP/UseCase2/'
FILE = 'BranchSalesDriver'
```

```python
cols = ["PPC Spend","Online Sales"]
df = pd.read_csv(PATH+FILE, usecols=cols)

x = df["PPC Spend"]
y = df["Online Sales"]
```

```python
corr = np.corrcoef(x,y, rowvar=False)
print("Correlation Coefficient is {:.2f}".format(corr[0,1]))
```

```python
# Run regression and get the different parameters
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
```

```python
print("p-value is {:.8f}".format(p_value))
print("r-squared is {:.8f}".format(np.square(r_value)))
print("Slope is {:.2f}".format(slope))

# 90% confidence interval: this many "standard errors" above and below \
# the regression slope
CI90 = 1.645
CI = std_err * CI90
slopeLow = slope - CI
slopeHigh = slope + CI
print("90% Confidence Interval is from {:.2f} to {:.2f}".format(slopeLow, slopeHigh))
```
