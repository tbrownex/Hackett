```python
import pandas as pd
import numpy  as np
from config import getClient
```

```python
dataloc = getClient("OutFront")
```

```python
df = pd.read_csv(dataloc+"2c_updatedHours.csv")

# Convert their date format to a standard date
df["date"] = pd.to_datetime(df["date"])

df = df.set_index(["panel", "date", "hour"])
```

```python
panel = input("enter the panel id")
dt    = input("enter the date in format YYYY-MM-DD")
```

```python
# View a "not found" panel/date
df.loc[panel, dt]
```

```python
hr  = int(input("enter the hour"))
pop = input("enter the population")
```

```python
# Set the value for the specified hour
df.loc[panel, dt, hr]["population"] = pop
```

```python
# Check that it worked
df.loc[panel, dt]
```
