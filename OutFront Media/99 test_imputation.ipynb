```python
# Check each panel for missing dates
df = df.set_index("panel")
grp = df.groupby(level=0)

count = 0
for idx, val in grp:
    missing = checkMissing(val, "day")
    if len(missing) > 0:
        count += 1
        print("found one missing: {}".format(idx))
print("Number missing dates: {}".format(count))
```

```python
def getMissing(grp):
    # Create list of panel/date with missing hours
    missingList = []
    for idx, val in grp:
        missing = val.loc[val["population"].isnull()]["hour"]
        if not missing.empty:
            missingList.append(idx)
    return missingList
```

```python
df  = df.set_index(["panel", "date"])
grp = df.groupby(level=[0,1])

missingList = getMissing(grp)

for x in missingList:
    panel = x[0]
    date  = x[1]
    tmp = df.loc[panel, date]["population"]
    clusters = tmp.isnull().astype(int).groupby(tmp.notnull().astype(int).cumsum()).sum()
    print(tmp)
    print(clusters)
    input()
```
