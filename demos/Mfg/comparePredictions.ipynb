```python
import random
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/tbrownex/predDF.csv")
```

```python
df.set_index("unit", inplace=True)
```

```python
df["pred"] = (df["RF"]+df["NN"]+df["XGB"])/3
```

```python
def plotComps(df, units, x=None):
    if x is None:
        x = np.random.choice(units, 1)
    df = df.loc[x]
    actuals = df["actual"]
    actuals = actuals.values.reshape([-1,1])
    predictions = df["pred"]
    predictions = predictions.values.reshape([-1,1])
    baseline = df["Baseline"]
    baseline = baseline.values.reshape([-1,1])
    plt.plot(actuals, label="actual")
    plt.plot(predictions, label="prediction")
    plt.plot(baseline, label="baseline")
    plt.legend(loc='upper right')
    unit = "Unit # " + str(x)
    plt.annotate(unit, xy=(0.05, 0.05), xycoords='axes fraction')
    plt.show()
```

```python
units = df.index.values
```

```python
while True:
    plotComps(df, units)
    input()
```
