```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def plotComps(df, x=None):
    if x is None:
        x = np.random.choice(np.unique(df.index.values))
    tmp = df.loc[x]
    actual     = np.array(tmp["actual"])
    prediction = np.array(tmp["0"])
    plt.plot(actual, label="actual")
    plt.plot(prediction, label="prediction")
    plt.legend(loc='upper right')
    unit = "Unit # " + str(x)
    plt.annotate(unit, xy=(0.05, 0.05), xycoords='axes fraction')
    plt.show()
```

```python
dataloc = "/home/tbrownex/"
file    = "df.csv"

df = pd.read_csv(dataloc+file)

df["diff0"]=df["actual"] - df["0"]

mape = abs(df["diff1"]/df["actual"]).mean()
print("MAPE: ", mape)

rmse = (df["diff1"]**(2)).mean()**(.5)
print("RMSE: ", rmse)
```
